**[Task 1.2 (d).]()** Binarize the image by applying an intensity threshold:

```python
T=0.1
mask_array = np.where(img_array >= T, 1,0)
plt.imshow(mask_array)
```

**[Task 1.2 (e).]()** Load the ground truth:

```python
mask_gd_array = plt.imread('data/NIH3T3/gt/0.png')
plt.imshow(mask_gd_array)
```

Compute and print the Dice coefficient:

```python
intersect = np.sum(mask_array * mask_gd_array)

sum = np.sum(mask_array) + np.sum(mask_gd_array)

dic_e = 2 * intersect / sum
print(dic_e)
```
