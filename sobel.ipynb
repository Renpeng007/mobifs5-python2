**[Task 2.2.]()** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical $3 \times 3$ Sobel filtering, respectively:

```python
def conv(img, kernel):
    k_h, k_w = kernel.shape
    i_h, i_w = img.shape
    p_h = k_h // 2
    p_w = k_w // 2
    padded_img = np.pad(img, ((p_h, p_h), (p_w, p_w)), mode='constant')
    output = np.zeros_like(img)

    for i in range(i_h):
        for j in range(i_w):
            region = padded_img[i:i+k_h, j:j+k_w]
            output[i,j] = np.sum(region * kernel)
    return output

def sobel_h(img):
    h_kernel = np.array([[1,0,-1], [2,0,-2], [1,0,-1]])
    cov_img = conv(img, h_kernel)
    return cov_img


def sobel_v(img):
    v_kernel = np.array([[1,2,1], [0,0,0], [-1,-2, -1]])
    cov_img = conv(img, v_kernel)
    return cov_img
```
